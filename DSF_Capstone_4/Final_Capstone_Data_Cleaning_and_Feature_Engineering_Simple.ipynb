```python
#since each hero can only be selected once per game, duplicate
#values cannot happen, so we can add all of the columns
#together to make a single set for each team
radiant_team = dummy1 + dummy2 + dummy3 + dummy4 + dummy5
dire_team = dummy6 + dummy7 + dummy8 + dummy9 + dummy0
column_names = []
for i in range(len(radiant_team.columns)):
    column_names.append('Radiant: ' + radiant_team.columns[i])
radiant_team.columns = column_names
column_names = []
for i in range(len(dire_team.columns)):
    column_names.append('Dire: ' + dire_team.columns[i])
dire_team.columns = column_names
```

```python
#and now we merge
df = df.merge(radiant_team, left_index=True, right_index=True)
df = df.merge(dire_team, left_index=True, right_index=True)
```

```python
#so, I don't make dummies of these as they are categorical numbers and I'm only
#interested in responses that are 2 or 3, but since I've gotten use to using
#dummy as the name of a df that I manipulate to eventually add to the main
#df, I named it such here
dummy_leaver = df[['first_leaver', 'second_leaver', 'third_leaver', 'fourth_leaver','fifth_leaver',  
                  'sixth_leaver', 'seventh_leaver', 'eighth_leaver', 'ninth_leaver', 'tenth_leaver']]
```

```python
#numpy's where function would have been more efficient here but I wanted to avoid
#using numpy, so I used 
dummy_leaver = dummy_leaver.mask(dummy_leaver < 2, 0)
```

```python
#since we only care about if a hero is on the team, we only care if someone on that
#team abandoned the game, so we'll treat this the same as the hero columns above
radiant_abandon = (dummy_leaver['first_leaver'] + dummy_leaver['second_leaver'] + 
                   dummy_leaver['third_leaver'] + dummy_leaver['fourth_leaver'] +
                   dummy_leaver['fifth_leaver'])
dire_abandon = (dummy_leaver['sixth_leaver'] + dummy_leaver['seventh_leaver'] + 
                dummy_leaver['eighth_leaver'] + dummy_leaver['ninth_leaver'] + 
                dummy_leaver['tenth_leaver'])
```

```python
#and here we convert any of these values to 1 so it matches the other binary columns
df['Radiant Abandon'] = radiant_abandon.mask(radiant_abandon >= 2, 1)
df['Dire Abandon'] = dire_abandon.mask(dire_abandon >= 2, 1)
```
